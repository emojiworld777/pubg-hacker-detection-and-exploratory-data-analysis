```python
def printinfo():
    temp = pd.DataFrame(index=df.columns)
    temp['dataType'] = df.dtypes
    temp['nullCount'] = df.isnull().sum()
    temp['uniqueCount'] = df.nunique()
    return temp
printinfo()
```

```python
x = df['solo_KillDeathRatio'].value_counts().index
y =df['player_name'].value_counts()
xaxis = []
yaxis = []
for i in range(len(x)):
    xaxis.append(x[i])
    yaxis.append(y[i])

plt.figure(figsize = (25,18))
plt.xlabel('solo_KillDeathRatio')
plt.ylabel('player_name')
graph = sns.barplot(x =xaxis,y = yaxis,palette ='husl')
graph.set_title('Top solo player based on there killdeath ration')
```

```python
sns.histplot(df['solo_KillDeathRatio'], kde = True)
plt.title('Kill counts Distribution')
plt.xlabel('Kill_Counts')
plt.ylabel('Counts')
plt.show()
```
