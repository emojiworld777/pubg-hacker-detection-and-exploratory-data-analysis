```python
def printinfo():
    temp = pd.DataFrame(index=df.columns)
    temp['dataType'] = df.dtypes
    temp['nullCount'] = df.isnull().sum()
    temp['uniqueCount'] = df.nunique()
    return temp
printinfo()
```

```python
x = df['solo_KillDeathRatio'].value_counts().index
y =df['player_name'].value_counts()
xaxis = []
yaxis = []
for i in range(len(x)):
    xaxis.append(x[i])
    yaxis.append(y[i])

plt.figure(figsize = (25,18))
plt.xlabel('solo_KillDeathRatio')
plt.ylabel('player_name')
graph = sns.barplot(x =xaxis,y = yaxis,palette ='husl')
graph.set_title('Top solo player based on there killdeath ration')
```

```python
sns.histplot(df['solo_KillDeathRatio'], kde = True)
plt.title('Solo Kill_death_ratio Distribution')
plt.xlabel('Kill_death_ratio solo')
plt.ylabel('Counts')
plt.show()
```

```python
# Solo Kill Distribution
kill = df['solo_Kills']
plt.hist(kill,bins = 100)
plt.xlabel("solo kill counts")
plt.ylabel("kill frequency")
plt.title('Solo Kill Distribution')
plt.show()
```

```python
# relation between kill and survival time
surv_time = df['solo_TimeSurvived']
corrx = np.corrcoef(kill,surv_time)[0,1]
print(corrx)
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap: Kill Counts vs. Survival Time")
plt.show()
```

```python
sns.jointplot(x= kill,y=surv_time,kind = 'hex')
plt.xlabel('kill counts')
plt.ylabel('survival time')
plt.title('Relation between Kills And Survival Time')
plt.show()
```
